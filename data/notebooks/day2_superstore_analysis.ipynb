```python
df_clean = df.dropna()
```

```python
df_clean.shape
```

```python
sales_by_category = df_clean.groupby("Category")["Sales"].sum().sort_values(ascending=False)
sales_by_category
```

```python
profit_by_region = df_clean.groupby("Region")["Profit"].sum().sort_values(ascending=False)
profit_by_region
```
